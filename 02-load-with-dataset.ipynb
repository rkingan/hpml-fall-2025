```python
from torch.utils.data import DataLoader, Dataset
import time

def time_dataloader(ds: Dataset, nw: int) -> float:
    dataloader = DataLoader(ds, batch_size=8, shuffle=False, num_workers=nw)
    start = time.monotonic()
    for i, _ in enumerate(dataloader):
        if i % 1000 == 0:
            print(f"At sample {i} / {len(dataloader)}")
    end = time.monotonic()
    return end - start

nw_values = [1, 2, 4, 8]
avg_times = []
trials = 4

for nw in nw_values:
    tot_time = 0.0
    log.info("Timing with %d workers...", nw)
    for trial in range(trials):
        ds = FaceLandmarksDataset(Path().resolve() / "facial_keypoints.csv", Path().resolve() / "images")
        elapsed = time_dataloader(ds, nw)
        log.info("...trial %d, elapsed time: %.2f seconds", trial, elapsed)
        tot_time += elapsed
    avg_times.append(tot_time / trials)
    log.info("Average time with %d workers: %.2f seconds", nw, avg_times[-1])
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(len(avg_times)), avg_times, marker='o')
ax.set_xticks(np.arange(len(nw_values)))
ax.set_xticklabels(nw_values)
ax.set_xlabel("Number of workers")
ax.set_ylabel("Average time (seconds)")
ax.set_title("DataLoader performance with different numbers of workers")
plt.show()
```
